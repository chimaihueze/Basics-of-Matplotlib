```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
# Display Multiple Plots
# With the subplots() function you can draw multiple plots in one figure:

# Example
# Draw 2 plots:

ab = np.array([1, 2, 3, 4, 8])
ac = np.array([10, 20, 30, 40, 80])

plt.subplot(1, 2, 2)
plt.plot(ab, ac)
```
